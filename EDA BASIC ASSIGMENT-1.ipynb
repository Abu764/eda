$ Add new column$

```python
name=['Rakesh','jitesh','anna']
age=[41,42,43]

data=zip(name,age)
cols=['name','age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
df
```

```python
name=['Ramesh','suresh','sathish']
age=[30,35,40]

df=pd.DataFrame(zip(name,age),
                columns=['name','age'],
                index=['A','B','C'])

city_names=['hyd','blr','chennai']
df['city']=city_names

df.drop('city',
        axis=1,
        inplace=True)
df
```

# step-8

$ Drop rows $

```python
df.drop('C',
        axis=0,
        inplace=True)
df
```

# step-9

$ save the dataframe$

```python
df.to_csv("output.csv")
df.to_excel("output.xlsx")
```
